## Reshape

```python
# 차원만 바꿔준다.
arr = np.array([[1, 2, 3], [1, 2, 3], [1, 2, 3]])
print(arr)
arr = arr.reshape([1,9])
print(arr.shape)
print(arr)
```

```python
arr = np.array([[1, 2, 3], [1, 2, 3], [1, 2, 3]])
arr = arr.reshape(-1)
print(arr)
print(arr.shape)
```

```python
arr = np.array([[1, 2], [1, 2], [1, 2]])
print(arr.shape)
arr = arr.reshape(2, -1)
print(arr)
print(arr.shape)
```

## np.expand_dims()
#### 안의 값은 유지하되 차원 수를 늘리고 싶을 때

```python
arr = np.random.randn(3 , 3)
arr.shape
```

```python
arr = arr.ravel()
print(arr)
print(arr.shape)
arr = np.expand_dims(arr, -1)
print(arr)
print(arr.shape)
```

# Graph Visualization

#### Basic Attribute
alpha : 투명도 <br>
kind : 그래프 종류 'line', 'bar', 'barh', 'kde' <br>
logy : Y축에 대해 Log Scaling <br>
use_index : 객체에 색인을 눈금 이름으로 사용할지 여부 <br>
rot : 눈금 이름 돌리기 (rotating) 0 ~ 360 <br>
xticks, yticks : x, y축으로 사용할 값 <br>
xlim, ylim : X, Y축의 한계 <br>
grid : 축의 그리드를 표현할지 여부 <br>
 <br>
subplots : 각 column에 독립된 subplot 그리기 <br>
sharex, sharey : subplots=True 이면 같은 X,Y축을 공유하고 눈금과 한계를 연결 <br>
figsize : 생성될 그래프의 크기를 tuple로 지정 <br>
title : 그래프의 제목 지정 <br>
legend : subplot의 범례 지정 <br>
sort_columns : column을 알파벳 순서로 그린다. <br>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randn(50).cumsum()
data
```

```python
plt.plot(data)
plt.show()
```

## 여러 그래프 그리기

```python
plt.subplot(2, 2, 1) # 그릴준비
plt.subplot(2, 2 ,2)
plt.subplot(2, 2, 3)
plt.show()
```

```python
hist_data = np.random.randn(100)
scat_data = np.arange(30)
```

```python
plt.subplot(2, 2, 1)
plt.plot(data)
plt.subplot(2, 2, 2)
plt.hist(hist_data, bins=100)
plt.subplot(2, 2, 3)
plt.scatter(scat_data, np.arange(30) * np.random.randn(30))
plt.show()
```

## 그래프 선 옵션
- 색상 : b(파란색), g(초록색), r(빨간색), c(청록색), y(노란색), k(검은색), w(흰색)
- 마커 : o(원), v(역삼각형), ^(삼각형), s(네모), +(플러스), .(점)

```python
plt.plot(data, 'ys')
plt.show()
```

## 그래프 사이즈 조절

```python
plt.figure(figsize=(10,3)) # subplot보다 맨 위에 있어야 한다.
plt.plot(data, 'k+')
plt.show()
```

## 그래프 겹치기 + legend 달기

```python
data = np.random.randn(30).cumsum()
plt.figure(figsize=(10,5))
plt.plot(data, label='default')
plt.plot(data, 'k--', drawstyle='steps-post',label='steps-post')
plt.legend()
plt.show()
```

## 이름달기

```python
plt.plot(np.random.randn(1000).cumsum())
plt.title("Random Graph")
plt.xlabel("Stages")
plt.ylabel("Values")
plt.show()
```

## 종합

```python
plt.title("TESTER")
plt.plot(np.random.randn(1000).cumsum(),
'k', label='one')
plt.plot(np.random.randn(1000).cumsum(),
'b', label='two')
plt.plot(np.random.randn(1000).cumsum(),
'r', label='three')
plt.legend()
plt.show()
```
